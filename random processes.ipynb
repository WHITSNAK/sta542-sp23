```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm

sns.set_style('whitegrid')
```

### Random Walk

```python
sigma = 5
df = pd.DataFrame(np.random.normal(loc=0, scale=sigma, size=(200, 1000))).T.cumsum()
```

```python
fig, ax = plt.subplots(figsize=(10, 5))

df.plot(ax=ax)

ax.get_legend().remove()
```

```python
df.quantile(0.025, axis=1).plot()
df.quantile(0.975, axis=1).plot()
df.mean(axis=1).plot()
```

### MA(1)

```python
sigma = 5
theta = 0.3
zs = np.random.normal(loc=0, scale=sigma, size=1000)
```

```python
plt.plot(zs[1:] + theta*zs[:-1])
```

```python
pd.Series(zs).rolling(2).apply(lambda x: x.values[1] + theta*x.values[0]).plot()
```

```python
pd.Series(zs).rolling(10).mean().plot()
```
