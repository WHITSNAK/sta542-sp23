```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from scipy import stats
from scipy.io import loadmat

sns.set_style('whitegrid')
```

## Ex 1

```python
sampling_freq = 200
eta = np.pi
ts = np.arange(1, 1000+1) / sampling_freq

xs = np.cos(2*np.pi * ts + eta)
gs = np.abs(xs)
```

```python
fig, ax = plt.subplots(ncols=2, figsize=(15, 4))

ax[0].plot(xs)
ax[1].plot(gs)

ax[0].set_title('$f(t)=cos(2\\pi t + \\beta)$')
ax[1].set_title('$g(t)=|f(t)|$')

fig.tight_layout()
fig.show()
```

```python
fig, ax = plt.subplots(ncols=2, figsize=(15, 4))

psd = np.abs(np.fft.fft(xs)) * (2/xs.shape[0])
ax[0].plot(psd[:20])

psd = np.abs(np.fft.fft(gs)) * (2/gs.shape[0])
ax[1].plot(psd[:50])

ax[0].set_title('$f(t)=cos(2\\pi t + \\beta)$')
ax[0].set_ylabel('Spectrum Density')
ax[0].set_xlabel('Frequency (Hz)')
ax[1].set_title('$g(t)=|f(t)|$')
ax[1].set_ylabel('Spectrum Density')
ax[1].set_xlabel('Frequency (Hz)')

fig.tight_layout()
fig.show()
```
