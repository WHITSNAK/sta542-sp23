```python
import sys

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from scipy import stats

sns.set_style('whitegrid')
```

```python
T = 1000
freq = 2*np.pi * 5 / T
wavelength = 2*np.pi/freq
rho = 1
eta = np.pi

ts = np.arange(0, T)
xs = rho * np.cos(freq*ts + eta)

plt.plot(ts, xs)
```

```python
wk = 2*np.pi/T
logl = np.log(np.abs(np.fft.fft(xs))**2/T*2)[:int(np.round(T, 0)/2)]
wavelengths = (2*np.pi) / (np.arange(1, int(np.round(T, 0)/2)+1) * wk)

fig, ax = plt.subplots(figsize=(8, 5))
ax.plot(wavelengths, logl)
ax.set_xlabel('Wave Length')
ax.set_ylabel('Log Likelihood')
```
