```python
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
phi = np.array([1.5, -0.75])
x0 = np.array([1, 1])
v = 1
A = np.array([phi, [1, 0]])

xs = []
x = x0
for i in range(144):
    w = np.array([np.random.standard_normal() * np.sqrt(v), 0])
    x = A @ x + w
    xs.append(x[0])
```

```python
def to_complex_polar(x):
    r = np.sqrt(x.real**2 + x.imag**2)
    theta = np.arctan(x.imag / x.real)
    return r, theta

lam, T = np.linalg.eig(A)
r, theta = to_complex_polar(lam)
wavelength = 2*np.pi / theta
```
